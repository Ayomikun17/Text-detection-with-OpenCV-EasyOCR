```python
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test1.png')
reader =easyocr.Reader(['en'], gpu = True)

image_text = reader.readtext(img)

threshold = 0.8

# Note: When the text in the image is read,the area where the text was seen in
#the image is specified alongside i.e 4 (x,y)coordinates

for t in image_text:
  print(t)
  bbox,text, score = t

  # Preventing the reader from reading noisy text by setting a threshold
  if score > threshold:

    # Drawing a bounding box around the text in the image
    cv2.rectangle(img, bbox[0], bbox[2], (0,255,0), 5)
    cv2.putText(img, text, bbox[0], cv2.FONT_HERSHEY_SIMPLEX, 0.65, (0,255,0), 2)

cv2.imwrite('output_image.jpg',img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
